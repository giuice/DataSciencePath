```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
star_wars.columns
```

```python
star_wars.head(10)
```

```python
star_wars.dropna(axis=0, subset=['RespondentID'], inplace=True)
star_wars.RespondentID.isnull().sum()
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
yes_no_map = {'Yes':True,
             'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = \
star_wars['Have you seen any of the 6 films in the Star Wars franchise?']\
    .map(yes_no_map)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = \
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']\
    .map(yes_no_map)
```

```python
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
```

```python
import numpy as np
star_wars[~pd.isnull(star_wars.iloc[:,3])]
```

```python
sw_columns = star_wars.columns[3:9]

for col in sw_columns:
    col_value = star_wars[~pd.isnull(star_wars[col])][col].iloc[0]
    col_name = col_value.replace("  "," ").\
             replace(" ", "_").\
             lower().\
             replace("star_wars:_episode","seen")
    star_wars[col] = star_wars[col].map({col_value:True, np.NaN:False})
    star_wars = star_wars.rename(columns={col:col_name})

star_wars.head()
    
```

```python
rank_columns  = star_wars.columns[9:15]
star_wars[rank_columns] = star_wars[rank_columns].astype(float)
count = 1
for col in rank_columns:
    star_wars = star_wars.rename(columns={col:'ranking_' + str(count)})
    count = count +1
rank_columns  = star_wars.columns[9:15]
star_wars[rank_columns].head()
```

```python
star_wars[rank_columns].mean()
```

## Qual a classificação por generos?
Finalmente para finalizar vamos analisar como diferem as classificações entre homens e mulheres a depois vamos ver a classificação do publico por genero, e finalmente ver a correlação entre que de quem gosta de star wars gosta de star trek?

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
fig.tight_layout()
(7 -males[rank_columns] ).mean().plot.bar(ax=axes[0])
(7 -females[rank_columns] ).mean().plot.bar(ax=axes[1])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
fig.tight_layout()
axes[0].bar(range(6), males[males.columns[3:9]].sum())
#plt.show()

axes[1].bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```
